```python
%matplotlib inline 
rcParams['figure.figsize'] = 8,4
plt.style.use('seaborn-whitegrid')
```

```python
address = '/Users/fyenne/Desktop/booooks/python/Python_course/Exercise Files/Data/mtcars.csv'
cars = pd.read_csv(address) #header = infer

cars.columns.values[0]='car_names' # rename first_position, col_name 
cars.head()
```

```python
mpg = cars.mpg
plt.plot(mpg)
```

```python
cars[['mpg']].describe()
```

```python
mpg_matrix = mpg.values.reshape(-1,1)
# mpg_matrix  # one line matrix

scaled = preprocessing.MinMaxScaler() # named function
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg) # y-axis are changed to between 0&1
```

```python
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))
scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg) # y-axis are changed to between 0 to 10
```

```python
test = preprocessing.MinMaxScaler().fit_transform(mpg_matrix)
plt.plot(test)
```

# Standardized

```python
standardized_mpg = scale(mpg, 
                         axis = 0, 
                         with_mean = True, 
                         with_std = False)
plt.plot(standardized_mpg)
```
